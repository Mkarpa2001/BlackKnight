# Data Cleaning

```python
cols_nan=['MasVnrType','BsmtQual','BsmtCond','BsmtFinType1','BsmtExposure','BsmtFinType2','Electrical','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond',]

for column in cols_nan:
    range_value=df[column].unique()
    len_value=len(df[column].unique())
    count=list(range(len_value))
    for value in df[column]:
        for i in range(len_value):
            if value== range_value[i]:
                count[i]+=1
    max_val=max(count)
    max_index=count.index(max_val)
    df[column].fillna(range_value[max_index],inplace=True)
    count=[0]*len_value
```

```python
df.isnull().sum()
```

```python
cols_del=['Alley','Id','GarageYrBlt','PoolQC','Fence','MiscFeature','YearBuilt']
df.drop(cols_del,axis=1,inplace=True)
```

```python
df.shape
```

```python
cols_to_dummy=df.select_dtypes(include=['object']).columns
```

```python
len(cols_to_dummy)
```

```python
dummies=pd.get_dummies(df[cols_to_dummy])
print(dummies.shape)
df=pd.concat([df,dummies],axis=1)
df.drop(cols_to_dummy,axis='columns',inplace=True)
```

```python
df.shape
```

```python
X=df.drop(['SalePrice'],axis=1)
y=df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import Ridge
lasso_reg=Ridge(alpha=0.1,max_iter=10000000,tol=0.00000001)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
kf=StratifiedKFold(n_splits=10,random_state=None)
max_score=0
for train_index,test_index in kf.split(X,y):
    X_train,X_test=X.iloc[train_index],X.iloc[test_index]
    y_train,y_test=y.iloc[train_index],y.iloc[test_index]
    lasso_reg.fit(X_train,y_train)
    if max_score < lasso_reg.score(X_test,y_test):
        pred=lasso_reg.predict(X_test)
        coef=lasso_reg.coef_
        intercept=lasso_reg.intercept_
        max_score=lasso_reg.score(X_test,y_test)
print(max_score) 
```

```python
coef
```

```python
def predict_price(items,coef,intercept):
    result=0
    for i in range(len(coef)):
        result+=items[i]*coef[i]
    return (result+intercept)
```

```python
predict_val=[]
for row in X_test.values:
    predict_val.append(predict_price(row,coef,intercept))
    
```
